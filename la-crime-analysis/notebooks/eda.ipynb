```python
## Visualize Crime Trends Over Time:

# Plot the number of crimes over time (e.g., by month or year).

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# # Group by month and count crimes
monthly_crimes = df.groupby('month').size()
```

```python
# Visualize the distribution of crime types

plt.figure(figsize=(12,6))
sns.countplot(y='crm_cd', data=df, order=df['crm_cd'].value_counts().index)
plt.title('Crime Distribution by Type')
plt.xlabel('Count')
plt.ylabel('Crime Type')
plt.savefig(r'C:\Users\Anitta\data-portfolio-projects\la-crime-analysis\visualizations\crime_distribution_by_type_countplot.png')
plt.show()
```

```python
# Plot

plt.figure(figsize=(12,6))
monthly_crimes.plot(kind='line')
plt.title('Monthly Crime Trends')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')

plt.savefig(r'C:\Users\Anitta\data-portfolio-projects\la-crime-analysis\visualizations\monthly_crime_trends.png')
plt.show()
```

```python
top_20_crime_codes = df['crm_cd'].value_counts().head(20)

plt.figure(figsize=(12,6))

sns.barplot(x=top_20_crime_codes.values, y=top_20_crime_codes.index, orient='h')

plt.title('Top 20 Most Frequent Crime Codes')
plt.xlabel('Count')
plt.ylabel('Crime Code')
plt.savefig(r'C:\Users\Anitta\data-portfolio-projects\la-crime-analysis\visualizations\crime_distribution_by_type_barplot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Map the integer values to day names (1=Monday, 2=Tuesday, ..., 7=Sunday)
day_of_week_mapping = { 
    1: 'Monday', 
    2: 'Tuesday', 
    3: 'Wednesday', 
    4: 'Thursday', 
    5: 'Friday', 
    6: 'Saturday',
    7: 'Sunday'
}

# Map the day_of_week column from integers to day names
df['day_of_week'] = df['day_of_week'].map(day_of_week_mapping)

# Now plot the data
plt.figure(figsize=(12, 6))
sns.countplot(x='day_of_week', data=df, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Crime by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Crimes')
plt.tight_layout()  # Adjust layout to make sure everything fits

# Save the plot as a .png file
plt.savefig(r'C:\Users\Anitta\data-portfolio-projects\la-crime-analysis\visualizations\crime_by_day_of_week_barplot.png')

# Show the plot
plt.show()
```
